```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
ps=PorterStemmer()
lemmatizerr=WordNetLemmatizer()
```

```python
sentences=nltk.sent_tokenize(paragraph_repeat)
```

```python
corpus=[]
```

```python
for i in range (len(sentences)):
    removal=re.sub('[^a-zA-Z]',' ',sentences[i])
    removal=removal.lower()
    removal=removal.split()
    removal=[lemmatizerr.lemmatize(word)for word in removal if not word in set(stopwords.words('english'))]
    removal=' '.join(removal)
    corpus.append(removal)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c_v=CountVectorizer(max_features=1500)
vector=c_v.fit_transform(corpus).toarray()
```

```python
df = pd.DataFrame(data=vector)
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf=TfidfVectorizer(max_features=1500)
tfidf_vector=tf_idf.fit_transform(corpus).toarray()
```

```python
# Cleaning and pre-processing
text_cleaning=re.sub(r'\[[0-9]]*\]',' ',paragraph_repeat)
text_cleaning=re.sub(r'\s+',' ',text_cleaning)
text_cleaning=text_cleaning.lower()
text_cleaning=re.sub(r'\d',' ',text_cleaning)
```

```python
text_cleaning
```

```python
sentences=nltk.sent_tokenize(text_cleaning)
```

```python
sentences
```

```python
tokenized_words=[nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
tokenized_words
```

```python
exclud_stopwords=[]
```

```python
# Loop for making list of words
for words in (tokenized_words):
    for word in  words:
        if word not in stopwords.words('english'):
            exclud_stopwords.append(word)  
```

```python
for i in range (len(tokenized_words)):
    tokenized_words[i]=[word for word in tokenized_words[i] if word not in stopwords.words('english')]
```

```python
gensim_words=Word2Vec(tokenized_words,min_count=1)
```
